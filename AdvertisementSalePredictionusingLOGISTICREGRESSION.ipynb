**Load Dataset
**

```python
dataset = pd.read_csv('data.csv')
print(dataset)
```

**Summarize the dataset**

```python
print(dataset.shape) #No of rows and columns
print(dataset.head(5))
```

**Segregating dataset into X(Input/Independent variable) and Y(Output/Dependent variable)**

```python
X=dataset.iloc[:, :-1].values
X
```

```python
Y=dataset.iloc[:, -1].values
Y
```

**Splitting data into train and test**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

**Feature Scaling**


```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
```

***Training Data***

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression() #Loading the algorithm
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

***Confusion matrix***

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)

print("Accuracy of the model: {0}%".format(accuracy_score(y_test, y_pred)*100))
```

***Predicting, whether new customer will buy or not***

```python
age = int(input("Enter new customer's age: "))
sal = int(input("Enter new customer's salary:"))
newCust = [[age,sal]]
result = model.predict(sc.transform(newCust))
print(result)
if result==1:
  print("Customer will Buy")
else:
  print("Customer won't buy")
```
